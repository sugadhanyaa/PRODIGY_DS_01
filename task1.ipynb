```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('population.csv')
```

```python
df
```

```python
df = df.drop(columns=['Rank'])
```

```python
df
```

```python
df = df.sort_values(by='Population', ascending=False)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 20))




axes[0].bar(df['State'], df['Population'], color='skyblue')
axes[0].set_title('Population by state')
axes[0].set_xlabel('State')
axes[0].set_ylabel('Population')
axes[0].tick_params(axis='x', rotation=45)



axes[1].bar(df['State'], df['Sex Ratio'], color='lightgreen')
axes[1].set_title('Sex Ratio by state')
axes[1].set_xlabel('state')
axes[1].set_ylabel('Sex Ratio (Males per 100 Females)')
axes[1].tick_params(axis='x', rotation=45)




axes[2].bar(df['State'], df['Literacy Rate (%)'], color='coral')
axes[2].set_title('Literacy Rate by State')
axes[2].set_xlabel('State')
axes[2].set_ylabel('Literacy Rate (%)')
axes[2].tick_params(axis='x', rotation=45)



```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.hist(df['Population'], bins=15, color='skyblue', edgecolor='black')
plt.title('Population Distribution')
plt.xlabel('Population (Millions)')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(df['Sex Ratio'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Sex Ratio Distribution')
plt.xlabel('Sex Ratio (Males per 100 Females)')
plt.ylabel('Frequency')


plt.subplot(1, 3, 3)
plt.hist(df['Literacy Rate (%)'], bins=15, color='lightcoral', edgecolor='black')
plt.title('Literacy Rate Distribution')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('Frequency')


plt.tight_layout()


plt.show()
```
